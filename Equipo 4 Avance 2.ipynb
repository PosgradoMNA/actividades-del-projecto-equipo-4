CRoss Industry Standard Process for Data Mining (CRISP-DM) es un proceso que sirve para las bases de la data mining. Tiene seis fases:

1.	Entender el negocio (Business understanding) – ¿Que necesitas del base de datos para el negocio? 
>a.	Se enfoca en entender los objetivos de los datos, así como los requerimientos que queremos obtener, nos sirve para obtener un plan preliminar Bloque con sangría


2.	Entender los datos (Data understanding) –¿Que tenemos y que necesitamos de esto?
> a.	Comenzamos con la recolección inicial de datos , después nos familiarizamos con los datos entendiendo cada campo, es interesante entender también que sub casos podemos analizar una ves entendiendo esto podemos formular un hipótesis


3.	Preparación de datos (Data preparation) – ¿Como organizamos los datos para ser modelados?
> a.Consiste en construir un set de datos apartir de datos crudos sin procesar 

4.  Modelado (Modeling) – ¿Qué técnicas de modelado se deben aplicar?
> a.El analista evalúa, selecciona y aplica las técnicas de modelado apropiadas. Dado que algunas técnicas como las redes neuronales tienen requisitos específicos con respecto a la forma de los datos. Puede haber un bucle de regreso aquí a la preparación de datos
5.	Evaluacion (Evaluación) – ¿Cuál modelo describe mejor los objetivos del negocio?
> a.El analista crea y elige modelos que parecen tener alta calidad en función de las funciones de pérdida que se seleccionaron. El analista los prueba para asegurarse de que puedan generalizar los modelos contra datos no vistos. Posteriormente, el analista también valida que los modelos cubran suficientemente todos los temas clave del negocio. El resultado final es la selección de los modelos campeones.
6.	Lanzamiento(Implementación) – ¿Cómo los stakeholders pueden acceder a la información?
>a.	En general, esto significará implementar una representación de código del modelo en un sistema operativo. Esto también incluye mecanismos para puntuar o categorizar nuevos datos no vistos a medida que surgen. El mecanismo debe utilizar la nueva información en la solución del problema empresarial original. Es importante destacar que la representación del código también debe incluir todos los pasos de preparación de datos que conducen al modelado. Esto garantiza que el modelo tratará los nuevos datos sin procesar de la misma manera que durante el desarrollo del modelo.
